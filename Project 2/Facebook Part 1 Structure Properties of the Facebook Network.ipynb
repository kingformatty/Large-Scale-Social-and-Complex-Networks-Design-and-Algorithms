# Question 1 A First Look at the Network

 1.1 Report the number of nodes and number of edges of the Facebook network.

```r
#read network (specify as undirected graph, or the diameter can not match)
fb_net=read_graph('facebook_combined.txt',directed=FALSE)
#count edge and node
e_num=ecount(fb_net)
v_num=vcount(fb_net)
cat('The amount of edge is ', e_num,'\n')
cat('The amount of vertex is ',v_num)
```

```r
plot(fb_net,main='Facebook Network',vertex.size=4,edge.arrow.size=.1,vertex.label=NA)
```

# Question 3 Plot the degree distribution of the facebook network and report average degree

```r
plot(degree_distribution(fb_net),xlab='Degree',ylab='Frequency',main='Degree Distribution of Facebook Network')
cat('Average Degree is ',mean(degree(fb_net)))
```

# Question 4 Plot the degree distribution of Question 3 in a log-log scale. Try to fit a line to the plot and estimate the slope of the line.

```r
#extract frequency and corresponding degree index and discard 0-value
dd=degree_distribution(fb_net)
index=which(dd!=0)
log_dd=log(dd)[index]
log_d=log(seq(1:length(dd)))[index]

#fit a linear model
model=lm(log_dd~log_d)
plot(log_d,log_dd,abline(model),main='Degree Distribution of Facebook Network in Log Scale')
cat('The slope of the fitting linear curve is ',coefficients(model)[2])
```
