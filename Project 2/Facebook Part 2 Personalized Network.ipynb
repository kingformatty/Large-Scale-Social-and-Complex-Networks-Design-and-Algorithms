# Question 5 Create a personalized network of the user whose ID is 1. How many nodes and edges does this personalized network have?

```r
#use function make_ego_graph
fb_net=read_graph('facebook_combined.txt',format='ncol',directed=FALSE)
node1_nets=make_ego_graph(fb_net,order=1,nodes='0')
node1_net=node1_nets[[1]]#extract graph element from list
```

```r
plot(node1_net, edge.arrow.size=.1, vertex.size=4, vertex.label=NA, main='node 1 and its neighbors')
cat('The number of nodes of this personalized network is ',vcount(node1_net),'\n')
cat('The number of edges of this personalized network is ',ecount(node1_net))
```

# QUESTION 6: What is the diameter of the personalized network? Please state a trivial upper and lower bound for the diameter of the personalized network.

```r
cat('The diameter of this personalized network is ',diameter(node1_net))
```
